# Data Cleaning & Preprocessing

```python
# Define a function to recode the age groups
def recode_age(age_code):
    if age_code <= 3:
        return 'Young Adult (18-39)'
    elif age_code <= 7:
        return 'Middle Aged (40-59)'
    else:
        return 'Senior (60+)'

# Apply the function to create a new 'Age_Group' column
df['Age_Group'] = df['Age'].apply(recode_age)

# Verify the recoding was successful
print(df['Age_Group'].value_counts())
```

# One-Sample T-Test
[In the Body Mass Index (BMI) scale, established by the World Health Organization (WHO) and other health bodies, the number 25 is a critical clinical threshold.]

Test if the mean BMI is different from 25.

```python
from scipy import stats

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(df['BMI'], popmean=25)

print("--- One-Sample T-Test Results (BMI vs. 25) ---")
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Mean BMI: {df['BMI'].mean():.2f}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. The mean BMI is significantly different from 25.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference.")
```

Null Hypothesis (H₀): μ = 25

Alternative Hypothesis (H₁): μ ≠ 25

Significance Level (α): 0.05

Results:

μ (Hypothesized Population Mean): 25

t-statistic (Calculated): t_calculated = 257.78

Critical t-value (from table): t_critical ≈ ±1.96 (For a large sample size, α=0.05, two-tailed)

p-value: p_value = 0.0000

Decision Rule:

Reject H₀ if: |t_calculated| > t_critical OR p-value < α

Conclusion:
Since |257.78| > 1.96 and 0.0000 < 0.05, we reject the null hypothesis (H₀).

There is sufficient evidence at the 0.05 significance level to conclude that the true population mean BMI is significantly different from 25. The sample mean of 28.38 suggests it is significantly greater than the WHO healthy-weight threshold.

# One-Way ANOVA with Post-Hoc Tukey Test
Test if mean BMI differs across your new Age_Group variable.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Separate the data into the three age groups
group_young = df[df['Age_Group'] == 'Young Adult (18-39)']['BMI']
group_mid = df[df['Age_Group'] == 'Middle Aged (40-59)']['BMI']
group_senior = df[df['Age_Group'] == 'Senior (60+)']['BMI']

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(group_young, group_mid, group_senior)

print("\n--- One-Way ANOVA Results (BMI by Age Group) ---")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. At least one group mean is significantly different.")

    # Perform Tukey's HSD Post-Hoc Test since ANOVA was significant
    print("\n--- Tukey's HSD Post-Hoc Test ---")
    tukey_results = pairwise_tukeyhsd(endog=df['BMI'], groups=df['Age_Group'], alpha=0.05)
    print(tukey_results)

else:
    print("Conclusion: Fail to reject the null hypothesis. No significant differences found.")
```

Groups: Young Adult (18-39), Middle Aged (40-59), Senior (60+)

Null Hypothesis (H₀): μ_young = μ_middle = μ_senior (All group means are equal)

Alternative Hypothesis (H₁): At least one group mean is significantly different

Significance Level (α): 0.05

Results:

F-statistic (Calculated): F_calculated = 415.3345

Critical F-value (from table): F_critical ≈ 3.00 (Approximate value for df_between=2, df_within=large, α=0.05)

p-value: p_value = 0.0000

Decision Rule for ANOVA:

Reject H₀ if F_calculated > F_critical or p-value < α

Conclusion for ANOVA:
Since F_calculated (415.3345) > F_critical (3.00) and p_value (0.0000) < α (0.05), we reject the null hypothesis (H₀). There is sufficient evidence at the 0.05 significance level to conclude that at least one age group has a significantly different mean BMI.
